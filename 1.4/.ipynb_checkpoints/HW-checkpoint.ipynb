1. Построение распределений

Сгенерировать выборку с распределением Рэлэя (https://ru.wikipedia.org/wiki/Распределение_Рэлея) используя numpy и scipy и построить на основе ее кривую плотности распределения на основе гистограммы из Matplotlib.

```python
def rel_pdf(rel_sigma, n, delta_x):
    pdf_x = []  # Координаты по оси абсцисс
    pdf_y = []  # Координаты по оси ординат
    for i in range(1, n):
        x = i * delta_x
        pdf_x.append(x)
        pdf_y.append(((2 * x) / rel_sigma) * math.exp((-x ** 2) / rel_sigma))
    return (pdf_x, pdf_y)
```

```python
x, y = rel_pdf(1, 100, 0.05)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, sps.rayleigh.ppf(x))
plt.show()
```

3. Поиск зависимостей

```python
headers = ['age', 'sex', 'bmi', 'children', 'smoker', 'region', 'charges']
```

```python
df = pd.read_csv('https://lhd8ts4trw.a.trbcdn.net/cdn/yusuciqe45/00_data_engineer/dataset_home.txt',
                names=headers)
```

```python
df.head()
```

```python
df['sex'] = df['sex'].map({'male':1, 'female':0})
```

```python
df['smoker'] = df['smoker'].map({'no':0, 'yes':1})
```

```python
df.region.unique()
```

```python
reg_dict = {
    'northwest':0,
    'northeast':1,
    'southwest':2,
    'southeast':3
}
```

```python
df['region'] = df['region'].map(reg_dict)
```

```python
df.head(10)
```

3.1 Корреляция

```python
sns.pairplot(df, kind='scatter', hue='smoker')
plt.show()
```

```python
smokers = df[df['smoker']==1]
sns.heatmap(smokers.corr())
plt.show()
```

```python
smokers.plot(kind='scatter', x='bmi', y='charges')
plt.show()
```

У курящего населения существует положительная корреляция между расходами и индексом массы тела

3.2 Зависимость

```python
reg0 = df[(df['region']==0) & (df['smoker']==0)]
```

```python
reg0.plot(kind='scatter', x='age', y='charges')
plt.show()
```

Зависимость у некурящего населения в регионе northwest, с возрастом растут расходы

3.3 Независимость

```python
no_smokers = df[df['smoker']==0]
no_smokers.plot(kind='scatter', x='age', y='bmi')
plt.show()
```

Индекс массы тела напрямую не зависит от возраста. Встречаются и молодые люди с высоким индексом и возрастные - с низким
